```python
import pandas as pd

# Load dataset
df = pd.read_csv('../data/shared_care_record.csv')

# Show data
print(df.head())
```

```python
import json

# Take first row (one patient)
patient = df.iloc[0]

# Create FHIR Patient resource
fhir_patient = {
    "resourceType": "Patient",
    "id": str(patient["patient_id"]),
    "gender": patient["gender"].lower(),
    "birthDate": "1980-01-01"
}

# Print nicely
print(json.dumps(fhir_patient, indent=2))
```

```python
import json

# Take first row (one patient)
patient = df.iloc[0]

# Create FHIR Patient resource
fhir_patient = {
    "resourceType": "Patient",
    "id": str(patient["patient_id"]),
    "gender": patient["gender"].lower(),
    "birthDate": "1980-01-01"
}

# Print nicely
print(json.dumps(fhir_patient, indent=2))
```

```python
# Create FHIR Observation resource

observation = {
    "resourceType": "Observation",
    "id": "obs-" + str(patient["patient_id"]),
    "status": "final",
    "category": [
        {
            "coding": [
                {
                    "system": "http://terminology.hl7.org/CodeSystem/observation-category",
                    "code": "vital-signs"
                }
            ]
        }
    ],
    "code": {
        "text": patient["observation"]
    },
    "subject": {
        "reference": "Patient/" + str(patient["patient_id"])
    },
    "valueQuantity": {
        "value": patient["obs_value"]
    }
}

import json
print(json.dumps(observation, indent=2))
```

```python
# Create FHIR Condition resource

condition = {
    "resourceType": "Condition",
    "id": "cond-" + str(patient["patient_id"]),
    "clinicalStatus": {
        "text": "active"
    },
    "code": {
        "text": patient["condition"]
    },
    "subject": {
        "reference": "Patient/" + str(patient["patient_id"])
    }
}

print(json.dumps(condition, indent=2))
```

```python
# Save FHIR resources to files

import os

# Create outputs folder if not exists
os.makedirs("../outputs", exist_ok=True)

# Save Patient
with open("../outputs/patient.json", "w") as f:
    json.dump(fhir_patient, f, indent=2)

# Save Observation
with open("../outputs/observation.json", "w") as f:
    json.dump(observation, f, indent=2)

# Save Condition
with open("../outputs/condition.json", "w") as f:
    json.dump(condition, f, indent=2)

print("All FHIR resources saved successfully.")
```
