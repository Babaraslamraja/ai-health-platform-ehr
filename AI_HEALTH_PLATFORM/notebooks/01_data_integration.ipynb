```python
df_b = pd.read_csv("../data/ehr_system_B.csv")

df_b.head()
```

```python
df_combined = pd.concat([df_a, df_b], ignore_index=True)

df_combined.head()
```

```python
df_combined.info()
df_combined.duplicated().sum()
df_combined['patient_id'].nunique()
```

```python
# Sort by latest encounter
df_combined['encounter_date'] = pd.to_datetime(df_combined['encounter_date'])
df_sorted = df_combined.sort_values(by='encounter_date')

# Keep latest record per patient
df_patient_latest = df_sorted.groupby('patient_id').last().reset_index()

df_patient_latest.head()
```

### Shared Care Record Creation

Data from multiple EHR systems was integrated to create a unified patient-level dataset.

For each patient, the most recent encounter was selected to simulate a shared care record, enabling a single, trusted view of patient data across systems.

This step demonstrates the core concept of a platform-based EHR architecture.

```python
df_patient_latest.to_csv("../outputs/shared_care_record.csv", index=False)
```

### Key Insight

Integrating data from multiple EHR systems revealed that patient information is distributed across sources, but can be unified into a single, reliable record.
